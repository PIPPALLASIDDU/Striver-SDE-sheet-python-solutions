# Rotate a Linked List
In this article, we will solve the problem: “Rotate a Linked List”

Problem Statement: Given the head of a linked list, rotate the list to the right by k places.

```python
#brute force - make end node head for each iteratio of k
def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
    if head is None or head.next is None:
        return head
    for i in range(k):
        temp=head
        while temp.next.next:
            temp=temp.next
        end=temp.next
        temp.next=None
        end.next=head
        head=end
    return head
```

```python
#own apprach - calculate the k mod len , give number of node to change , then traverse through whole list and 
#find the new head and end nodes then change last node next to present head , new end next to null and atlast head to new head
def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
    if head is None or head.next is None:
        return head
    temp=head
    cnt=0
    while temp:
        cnt+=1
        temp=temp.next
    k=k%cnt
    if k==0:
        return head
    temp=head
    while temp:
        cnt-=1
        if cnt==k:
            end=temp
            st=temp.next
        prev=temp
        temp=temp.next
    prev.next=head
    head=st
    end.next=None
```

```python
#striver optimal approach
def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
    if head is None or head.next is None:
        return head
    temp=head
    cnt=1
    while temp.next:
        cnt+=1
        temp=temp.next
    temp.next=head                    #tail pointing to head (circular linkedlist)
    k=k%cnt
    end=cnt-k
    while end>0:
        temp=temp.next
        end-=1
    head=temp.next
    temp.next=None
    return head
```

# 3 Sum : Find triplets that add up to a zero
Problem Statement: Given an array of N integers, your task is to find unique triplets that add up to give a sum of zero. In short, you need to return an array of all the unique triplets [arr[a], arr[b], arr[c]] such that i!=j, j!=k, k!=i, and their sum is equal to zero.

```python
#brute force
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        from itertools import combinations
        com=combinations(nums,3)
        ans=[]
        for i in com:
            if sum(i)==0:
                x=list(sorted(i))
                if x in ans:
                    pass
                else:
                    ans.append(x)
        return ans
```

```python
#optimized approach
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        ans=[]
        for i in range(0,len(nums)-2):
            if i==0 or (i>0 and nums[i]!=nums[i-1]):
                lo=i+1
                hi=len(nums)-1
                sm=0-nums[i]
                while lo<hi:
                    if nums[lo]+nums[hi]==sm:
                        if sorted([i,hi,lo]) not in ans:
                            ans.append([i,hi,lo])
                        while lo<hi and nums[lo]==nums[lo+1]:
                            lo+=1
                        while lo<hi and nums[hi]==nums[hi-1]:
                            hi-=1
                        lo+=1
                        hi-=1
                    elif nums[lo]+nums[hi]<sm:
                        lo+=1
                    else:
                        hi+=1
        return ans
                    
```

# Trapping Rainwater

Problem Statement: 
- Given an array of non-negative integers representation elevation of ground. Your task is to find the water that can be trapped after rain.

```python
#brute force
def trap(self, height: List[int]) -> int:
    n=len(height)
    water=0
    for i in range(n):
        j=i
        left=0
        right=0
        while j>=0:
            left=max([left,height[j]])
            j-=1
        j=i
        while j<n:
            right=max([right,height[j]])
            j++
        water+=(min((right,left))-height[i])
    return water
```

```python
#Better solution - taking 2 arrays suffix arrays and prefix arrays to computer leftmax and right max
def trap(self, height: List[int]) -> int:
    n=len(height)
    water=0
    prefix,suffix=[],[]
    for i in range(1,n):
        prefix[i]=max([prefix[i-1],height[i]])
    for i in range(n-2,-1,-1):
        suffix[i]=max([suffix[i+1],height[i]])
    for i in range(n):
        water+=min([prefix[i],suffix[i]])-height[i]
    return water
    
```

```python
#Best solution - 2 pointer approach 
def trap(self, height: List[int]) -> int:
    n=len(height)
    water=0
    left,right=0,n-1
    maxleft,maxright=0,0
    while left<=right:
        if height[left]<=height[right]:
            if height[left]>=maxleft:
                maxleft=height[left]
            else:
                water+=(maxleft-height[left])
            left+=1
        else:
            if height[right]>=maxright:
                maxright=height[right]
            else:
                water+=(maxright-height[right])
            right-=1
    return water
```

# Remove Duplicates in-place from Sorted Array
Problem Statement:
- Given an integer array sorted in non-decreasing order, remove the duplicates in place such that each unique element appears only once. The relative order of the elements should be kept the same.

If there are k elements after removing the duplicates, then the first k elements of the array should hold the final result. It does not matter what you leave beyond the first k elements.

Note: Return k after placing the final result in the first k slots of the array.

```python
#my approach in python , using list which is mutable(can be expandanded and contracted ) ,
#traverse from back and delete the repeateed node , so at last list will be left with unique nodes and in order
def removeDuplicates(self, nums: List[int]) -> int:
        n=len(nums)
        x=[]
        for i in range(n-1,-1,-1):
            if nums[i] in x:
                del nums[i]
            else:
                x.append(nums[i])
        return len(nums)
```

```python
#optimal 2 pointer approach
def removeDuplicates(self, nums: List[int]) -> int:
    i=0
    for j in range(1,len(nums)):
        if nums[i]!=nums[j]:
            i+=1
            nums[i]=nums[j]
    return i+1
```

# Count Maximum Consecutive One’s in the array
Problem Statement: 
- Given an array that contains only 1 and 0 return the count of maximum consecutive ones in the array.

```python
#OPTIMAL APPROACH - single pointer approach as only 2 values 0 & 1 are allowed 
def findMaxConsecutiveOnes(self, nums: List[int]) -> int:
    mx=0
    cnt=0
    for i in range(len(nums)):
        if nums[i]==1:
            cnt+=1
            if mx<cnt:
                mx=cnt
        else:
            cnt=0
    return mx
```
