# Rotate a Linked List
In this article, we will solve the problem: “Rotate a Linked List”

Problem Statement: Given the head of a linked list, rotate the list to the right by k places.

```python
#brute force - make end node head for each iteratio of k
def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
    if head is None or head.next is None:
        return head
    for i in range(k):
        temp=head
        while temp.next.next:
            temp=temp.next
        end=temp.next
        temp.next=None
        end.next=head
        head=end
    return head
```

```python
#own apprach - calculate the k mod len , give number of node to change , then traverse through whole list and 
#find the new head and end nodes then change last node next to present head , new end next to null and atlast head to new head
def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
    if head is None or head.next is None:
        return head
    temp=head
    cnt=0
    while temp:
        cnt+=1
        temp=temp.next
    k=k%cnt
    if k==0:
        return head
    temp=head
    while temp:
        cnt-=1
        if cnt==k:
            end=temp
            st=temp.next
        prev=temp
        temp=temp.next
    prev.next=head
    head=st
    end.next=None
```

```python
#striver optimal approach
def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
    if head is None or head.next is None:
        return head
    temp=head
    cnt=1
    while temp.next:
        cnt+=1
        temp=temp.next
    temp.next=head                    #tail pointing to head (circular linkedlist)
    k=k%cnt
    end=cnt-k
    while end>0:
        temp=temp.next
        end-=1
    head=temp.next
    temp.next=None
    return head
```

# 3 Sum : Find triplets that add up to a zero
Problem Statement: Given an array of N integers, your task is to find unique triplets that add up to give a sum of zero. In short, you need to return an array of all the unique triplets [arr[a], arr[b], arr[c]] such that i!=j, j!=k, k!=i, and their sum is equal to zero.

```python
#brute force
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        from itertools import combinations
        com=combinations(nums,3)
        ans=[]
        for i in com:
            if sum(i)==0:
                x=list(sorted(i))
                if x in ans:
                    pass
                else:
                    ans.append(x)
        return ans
```

```python
#optimized approach
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        ans=[]
        for i in range(0,len(nums)-2):
            if i==0 or (i>0 and nums[i]!=nums[i-1]):
                lo=i+1
                hi=len(nums)-1
                sm=0-nums[i]
                while lo<hi:
                    if nums[lo]+nums[hi]==sm:
                        if sorted([i,hi,lo]) not in ans:
                            ans.append([i,hi,lo])
                        while lo<hi and nums[lo]==nums[lo+1]:
                            lo+=1
                        while lo<hi and nums[hi]==nums[hi-1]:
                            hi-=1
                        lo+=1
                        hi-=1
                    elif nums[lo]+nums[hi]<sm:
                        lo+=1
                    else:
                        hi+=1
        return ans
                    
```
