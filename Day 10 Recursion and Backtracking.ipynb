# Print All Permutations of a String/Array
Problem Statement: 
- Given an array arr of distinct integers, print all permutations of String/Array.

```python
#finding all permutations using a frequency array
class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        def func(nums,freq,ds,ans):
            if len(ds)==len(nums):
                ans.append(tuple(nums))
                return
            for i in range(0,len(nums)):
                if !freq[i]:
                    freq[i]=True
                    ds.append(nums[i])
                    func(nums,freq,ds,ans)
                    ds.pop()
                    freq[i]=False
            
        ans,ds=[],[]
        freq=[None]*len(nums)
        func(nums,freq,ds,ans)
        return ans
```

```python
#optimised approach
class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        def func(ind,nums,ans):
            if ind==len(nums):
                ans.append(tuple(nums))
                return
            for i in range(ind,len(nums)):
                nums[i],nums[ind]=nums[ind],nums[i]
                func(ind+1,nums,ans)
                nums[i],nums[ind]=nums[ind],nums[i]
            
        ans=[]
        func(0,nums,ans)
        return ans
```

# N Queen Problem | Return all Distinct Solutions to the N-Queens Puzzle
Problem Statement: 
- The n-queens is the problem of placing n queens on n × n chessboard such that no two queens can attack each other. Given an integer n, return all distinct solutions to the n -queens puzzle. Each solution contains a distinct boards configuration of the queen’s placement, where ‘Q’ and ‘.’ indicate queen and empty space respectively.


```python
class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        def solve(col,board,ans):
            if col==len(board):
                ans.append(cons(board.copy()))
                return
            for row in range(len(board)):
                if safe(board,row,col):
                    board[row][col]='Q'
                    solve(col+1,board,ans)
                    board[row][col]='.'
    
        def safe(board,row,col):
            duprow = row
            dupcol = col
            while (row >= 0 and col >= 0):
                if (board[row][col] == 'Q'):
                    return False
                row-=1
                col-=1


            row = duprow
            col = dupcol
            while (col >= 0):
                if (board[row][col] == 'Q'):
                    return False
                col-=1

            row = duprow
            col = dupcol
            while (col >= 0 and row < len(board)): 
                if (board[row][col] == 'Q'):
                    return False
                col-=1
                row+=1
            return True
        def cons(board):
            x=[]
            for i in board:
                x.append("".join(i))
            return x
        board=[['.' for i in range(n)] for j in range(n)]
        ans=[]
        solve(0,board,ans)
        return ans
```

# Sudoku Solver
Problem Statement:

- Given a 9×9 incomplete sudoku, solve it such that it becomes valid sudoku. Valid sudoku has the following properties.

         1. All the rows should be filled with numbers(1 – 9) exactly once.

         2. All the columns should be filled with numbers(1 – 9) exactly once.

         3. Each 3×3 submatrix should be filled with numbers(1 – 9) exactly once.

```python
class Solution:
    def solveSudoku(self, board: List[List[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        def valid(board,i,j,c):
            for ind in range(len(board)):
                if board[i][ind]==str(c):
                    return False
                if board[ind][j]==str(c):
                    return False
                if board[int(3*(i//3)+i//3)][int(3*(j//3)+i%3)]==str(c):
                    return False
            return True 
        

        def solve(board):
            for i in range(9):
                for j in range(9):
                    if board[i][j]=='.':
                        for c in range(1,10):
                            if valid(board,i,j,c):
                                board[i][j]=str(c)
                                if solve(board):
                                    return True
                                board[i][j]='.'
                        return False
            return True
        solve(board)
```

```python
class Solution:
	def solveSudoku(self, board: List[List[str]]) -> None:
		EMPTY = "."
		
		def solve() -> bool:
			# traverse the board one entry at a time, updating empty entries with a
			# potential value, and backtracking if the updated board isn't valid
			for i in range(9):
				for j in range(9):
					if board[i][j] == EMPTY:
						# entry is empty, try each value (1-9) and see if the updated board
						# is still valid
						for n in range(1, 10):
							if is_valid(i, j, str(n)):
								# n doesn't violate any constraints, update the board
								board[i][j] = str(n)
								
								# check if the updated board is valid
								if solve():
									# updated board is valid
									return True
								
								# updated board is invalid
								board[i][j] = EMPTY
								
						# did not find any n that were valid, backtrack
						return False
			return True
		
		def is_valid(row: int, col: int, entry: str) -> bool:
		"""Return True if no constraints are violated, False otherwise."""
			if any(
				# check if entry violates the row constraint
				board[row][i] == entry or
				# check if entry violates the column constraint
				board[i][col] == entry or
				# check if entry violates the subgrid constraint
				board[(row//3)*3+i%3][(col//3)*3+i//3] == entry
				for i in range(9)
			):
				return False
			return True
		
		solve()
```

# M – Coloring Problem

Problem Statement: 
- Given an undirected graph and a number m, determine if the graph can be colored with at most m colors such that no two adjacent vertices of the graph are colored with the same color.

```python
def graphColoring(graph, k, V):
    def safe(node,graph,color,n,col):
        for it in range(n):
            if graph[node][it]==1 and color[it]==col:
                return False
        return True
    def solve(node,graph,color,n,m):
        if node==n:
            return True
        for i in range(1,m+1):
            if safe(node,graph,color,n,i):
                color[node]=i
                if solve(node+1,graph,color,n,m):
                    return True
                color[node]=0
        return False
    color=[0]*V
    return solve(0,graph,color,V,k)
```

# Rat in a Maze
Rat in a Maze

- Consider a rat placed at (0, 0) in a square matrix of order N * N. It has to reach the destination at (N – 1, N – 1). Find all possible paths that the rat can take to reach from source to destination. The directions in which the rat can move are ‘U'(up), ‘D'(down), ‘L’ (left), ‘R’ (right). Value 0 at a cell in the matrix represents that it is blocked and the rat cannot move to it while value 1 at a cell in the matrix represents that rat can travel through it.

Note: In a path, no cell can be visited more than one time.

Print the answer in lexicographical(sorted) order

```python
class Solution:
    def findPath(self, m, n):
        if m[0][0]==0:
            return [-1]
        def solve(i,j,di,dj,arr,n,ans,move,vis):
            if i==n-1 and j==n-1:
                print(move)
                ans.append(move.copy())
                return
            dir="DLRU"
            for ind in range(4):
                nexti=i+di[ind]
                nextj=j+dj[ind]
                if (nexti>=0 and nextj>=0 and nexti<n and nextj<n) and (not vis[nexti][nextj] and arr[nexti][nextj]==1):
                    vis[i][j]=1
                    solve(nexti,nextj,di,dj,arr,n,ans,move+dir[ind],vis)
                    vis[i][j]=0

        
        di=[1,0,0,-1]
        dj=[0,-1,1,0]
        ans=[]
        move=""
        vis=[[0]*n]*n
        solve(0,0,di,dj,m,n,ans,move,vis)
        return ans
```

```python
class Solution:
    def findPath(self, m, n):
        # code here
        self.res = []
        if m[0][0] == 0:
            return [-1]
            
        def check(ans,i,j,mat,N):
            if i < 0 or i >= N or  j < 0 or j >= N:
                return False
                
            if mat[i][j] == 0 or mat[i][j] == 2:
                return False

            if i == N-1 and j == N-1:
                self.res.append(ans)
                return 
        
            mat[i][j] = 2 # visited 
            
            check(ans + 'U',i-1,j,mat,N)
            check(ans + 'R',i,j+1,mat,N)
            check(ans + 'D',i+1,j,mat,N)
            check(ans + 'L',i,j-1,mat,N)
            
            mat[i][j] = 1 # unvisit it 
                   
        check('',0,0,m,n)
        if not self.res:
            return [-1]
        
        return self.res
```
